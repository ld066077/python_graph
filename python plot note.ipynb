```python
import matplotlib.pyplot as plt
```

## subgraph

在Matplotlib中，subplot函数用于在同一个图形窗口中创建多个子图。参数subplot(nrows, ncols, index)定义了子图的布局和位置。nrows和ncols分别表示子图的行数和列数，index表示子图的位置编号，编号从左到右，从上到下按顺序增加。

```python
ax1 = plt.subplot(212)
ax1.plot()
ax1.text(0.5,0.5,'212',horizontalalignment='center', verticalalignment='center',transform=ax1.transAxes, fontsize=20)

ax2 = plt.subplot(221)
ax2.plot()
ax2.text(0.5,0.5,'221',horizontalalignment='center', verticalalignment='center',transform=ax2.transAxes, fontsize=20)

ax3 = plt.subplot(222)
ax3.plot()
ax3.text(0.5,0.5,'222',horizontalalignment='center', verticalalignment='center',transform=ax3.transAxes, fontsize=20)

plt.show()
```
